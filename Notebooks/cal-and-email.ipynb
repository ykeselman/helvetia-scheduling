```python
from client_calendar import DavClient, SCalendar, SEvent, InfomDavClient, SogoDavClient
from client_email import EmailClient
```

```python
email = EmailClient()
invites = email.invites(move=False)
print("GOT INVITES", len(invites))

dav = SogoDavClient()
for invite in invites:
    sender = invite.sender
    invite = invite.cal_accepted
    cal = SCalendar(dav, sender)
    # cal.remove()
    event = cal.add_event(invite)
    print(event)
```
